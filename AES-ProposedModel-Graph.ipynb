```python
import pandas as pd
excel_file_path = 'D:\YahyaData\OutPut\Analysis\AES-Comparison.xlsx'

# Read the Excel file into a pandas DataFrame
df = pd.read_excel(excel_file_path)

# Display the DataFrame (optional)
print(df)
```

```python
import matplotlib.pyplot as plt
message = df['Message']

# Plot the data
plt.figure(figsize=(6, 4))
plt.plot(message, df['AES-Model'], label='AES')
plt.plot(message, df['Proposed-Model'], label='Proposed-Model')

# Customize the chart
plt.title('Comparison of AES and Proposed Model on Length')
plt.xlabel('Message')
plt.ylabel('Values')
plt.legend()
plt.grid(True)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the mean values for 'AES' and 'Proposed Model'
mean_salsa20 = df['AES-Model'].mean()
mean_proposed_model = df['Proposed-Model'].mean()

# Plot the bar chart
plt.figure(figsize=(3, 4))
plt.bar(['AES-Model', 'Proposed-Model'], [mean_salsa20, mean_proposed_model], color=['blue', 'orange'])
plt.title('Average Values Comparison: AES vs. Proposed Model on mean values')
#plt.xlabel('Model')
plt.ylabel('Average Values')
plt.show()
```
